```python
import re
import pandas as pd
import requests

pd.set_option('display.max_columns', None) # Para poder visualizar todas las columnas


API_KEY = "127708d4-cd8f-4bf7-89c5-1f6260459ad7"
URL_JUGADORES = "https://api.balldontlie.io/v1/players"

# Encabezados de la solicitud, incluyendo la clave API para autorización
headers = {
    'Authorization': API_KEY
}

# Realizar la solicitud GET
response = requests.get(URL_JUGADORES, headers=headers)
jugadores = response.json()

print(f"Si imprimo la llamada jugadores obtengo: {response}")
print(f"El estatus code de la llamada ha sido: {response.status_code}")
print(jugadores)
print(type(jugadores))

# Jugadores es un DICCIONARIO con las siguientes claves: dict_keys(['data', 'meta'])

print (f"El diccionario de jugadores tiene {len(jugadores.keys())} claves, que son:")
print(jugadores.keys())

# Nos adentramos un nivel: 

jugadores["data"]

# He accedido a una LISTA donde cada elemento es el diccionario de un jugador. 

print(f"La lista de jugadores['data'] tiene una longitud de {len(jugadores["data"])}")

```

```python

# Vuelvo a realizar la solicitud GET, esta vez indicamos que queremos el máximo de resultados por página: 100

# Parámetros de la solicitud para obtener una cantidad grande de jugadores
params = {
    'page': 1,       # Página de resultados
    'per_page': 100  # Cantidad de resultados por página
}


# Realizar la solicitud GET
response = requests.get(URL_JUGADORES, params=params, headers=headers)
jugadores = response.json()


print(f"Si imprimo la llamada jugadores obtengo: {response}")

# Jugadores es un DICCIONARIO con las siguientes claves: dict_keys(['data', 'meta'])

print (f"El diccionario de jugadores tiene {len(jugadores.keys())} claves, que son:")
print(jugadores.keys())

# Nos adentramos un nivel: 

jugadores["data"]

# He accedido a una LISTA donde cada elemento es el diccionario de un jugador. 

print(f"La lista de jugadores['data'] tiene una longitud de {len(jugadores["data"])}")
```
