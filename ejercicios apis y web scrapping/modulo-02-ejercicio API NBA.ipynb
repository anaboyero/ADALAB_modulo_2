```python
# PARTE 1 - Sacamos información de los jugadores.

import pandas as pd
import requests

pd.set_option('display.max_columns', None) # Para poder visualizar todas las columnas


nombres = []
apellidos = []
equipos = []
diccionario_jugadores = {'nombres': nombres, 'apellidos': apellidos, "equipos": equipos}

API_KEY = "127708d4-cd8f-4bf7-89c5-1f6260459ad7"
ENDPOINT_JUGADORES = "https://api.balldontlie.io/v1/players"

# Hago la llamada. Meto la clave. Compruebo si la llamada se ha hecho correctamente. 
# En ese caso me descargo el contenido de JSON, lo exploro y voy metiéndolo en un diccionario a medida que lo itero.

respuesta = requests.get(ENDPOINT_JUGADORES, headers= {'Authorization': '127708d4-cd8f-4bf7-89c5-1f6260459ad7'})
if respuesta.status_code != 200:
    print(f"Error en la llamada. Motivo: {respuesta.reason}")
else:
    print("Llamada correcta")
    jugadores = respuesta.json()
    num_jugadores = len(jugadores['data'])

    print(f"Cuántos jugadores hay: {num_jugadores}")

    for jugador in jugadores['data']:
        nombre = jugador['first_name']
        apellido = jugador['last_name']
        equipo = jugador['college']

        nombres.append(nombre)
        apellidos.append(apellido)
        equipos.append(equipo)

    # Y ahora hago un bonito dataframe con este diccionario

jugadores_NBA = pd.DataFrame(diccionario_jugadores)
jugadores_NBA

```

```python
# PARTE 2 - Sacamos información de los equipos.

conferencias = []
ciudades = []
nombres = []
abreviaciones = []

ENDPOINT_EQUIPOS = "https://api.balldontlie.io/v1/teams"
diccionario_equipos = {'conferencia': conferencias, 'ciudad': ciudades, 'nombre': nombres, 'abreviación': abreviaciones}


respuesta2 = requests.get(ENDPOINT_EQUIPOS, headers= {'Authorization': '127708d4-cd8f-4bf7-89c5-1f6260459ad7'})

if respuesta2.status_code != 200:
    print(f"Error en la llamada. Motivo: {respuesta2.reason}")
else:
    print("Llamada correcta")
    equipos = respuesta2.json()
    for equipo in equipos['data']:
        conferencia = equipo['conference']
        ciudad = equipo['city']
        nombre = equipo['name'] 
        abreviacion = equipo['abbreviation']

        conferencias.append(conferencia)
        ciudades.append(ciudad)
        nombres.append(nombre)
        abreviaciones.append(abreviacion)

    
    # Creo mi DataFrame con el diccionario actualizado
    equipos_NBA = pd.DataFrame(diccionario_equipos) 
    print(equipos_NBA.to_markdown())
```
