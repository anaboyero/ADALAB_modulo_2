Utiliza la variable content para crear la sopa. En la sintaxis de la sopa usa la variable content cómo si fuera response.content

```python
sopa = BeautifulSoup(content, 'html.parser')

# Inicializo siete listas que representan los valores de las columnas de mi futuro dataframe de rutas de montaña

lista_nombre = []
lista_descripcion = []
lista_lugar = []
lista_tipo = []
lista_duracion = []
lista_kilometros = []
lista_dificultad = []
lista_esfuerzo = []

# Extraemos los 10 nombres de las rutas

elemento_nombre = sopa.find_all('h3', {'class': 'elementor-heading-title elementor-size-default'})
for elemento in elemento_nombre[2:]: # Agregamos los nombres de las rutas. Hemos ignorado los primeros elementos, que no corresponden a nombres de rutas
    nombre_ruta = elemento.text
    nombre_ruta = nombre_ruta.replace("\n", ",")
    lista_nombre.append(nombre_ruta)    
    # print(f"entrando en nombre ruta {nombre_ruta}")

# Extraemos el lugar, el tipo, la duración, esfuerzo y dificultad técnica. Todos estos datos se encuentran en el mismo elemento.

elemento_lista = sopa.find_all('ul')
# print("longitud de elemento_listas", len(elemento_lista)) # Hay 17 elementos lista

for lista in elemento_lista[0:12:2]:    # Este elemento incluye más listas de las que necesitamos. Filtramos solo las que nos interesan. 
    apartados = lista.find_all('li')
    
    # Extraigo los datos que necesito 

    mi_lugar = apartados[0].text
    mi_tipo = apartados[1].text
    mi_duracion = apartados[2].text 
    mi_longitud = apartados[3].text 
    mi_dificultad = apartados[4].text 
    mi_esfuerzo = apartados[5].text 

    # Limpio los datos antes de incorporarlos a mis listas. Elimino los titulares y unidades y me quedo solo con los valores. Elimino espacios. Convierto los kilómetros a float.

    lista_lugar.append(mi_lugar.replace("Donde se realizan:", "").replace("\xa0", ""))  
    lista_tipo.append(mi_tipo.replace("Tipo de ruta:", "").replace('\xa0', ""))  
    lista_duracion.append(mi_duracion.replace("Duración:", "").replace("\xa0",""))  
    lista_kilometros.append(float(mi_longitud.replace("Km: ","").replace("km: ", "")))
    lista_dificultad.append(mi_dificultad)
    lista_esfuerzo.append(mi_esfuerzo.replace("Esfuerzo:", "").replace('\xa0', ""))


    # Extraigo las descripciones

    elemento_toggle = sopa.find_all('div', {'class': 'elementor-toggle'})
    # print(f"He encontrado {len(elemento_toggle)} elementos toggles")

    for toggle in elemento_toggle:
        # print("entro en toggle")
        apartados = toggle.find_all('div', {'class':'elementor-toggle-item'})
        # La información que me interesa está en el tercer apartado 
        apartado_descripcion = apartados[2]
        # print("len apartado descripcion:")
        # len(apartado_descripcion)
        mi_descripcion = apartado_descripcion.find('p').text
        lista_descripcion.append(mi_descripcion)

```

```python
# Ya tengo todos los datos (a falta de limpiarlos). Podría hacer el dataframe ya:

lista_rutas = []

lista_rutas2 = [lista_nombre, lista_descripcion, lista_lugar, lista_tipo, lista_duracion,lista_kilometros, lista_dificultad, lista_esfuerzo]

df1 = pd.DataFrame(lista_rutas2)
print(df1.to_markdown())
```
